```python
import pandas as pd

# Load dataset
df = pd.read_csv('Population of India.csv')

# Check original column names
print("Original columns:", df.columns.tolist())

# Clean column names by removing extra characters and spaces
df.columns = df.columns.str.replace(r'\[.*\]', '', regex=True).str.strip()
df.rename(columns={
    'Area (km2)': 'Area_km2',
    'Density (per km2)': 'Density'
}, inplace=True)

# Keep only necessary columns
columns_to_keep = ['State/UT', 'Population', 'Male', 'Female', 'Rural', 'Urban', 'Area_km2', 'Density']
df_cleaned = df[columns_to_keep]

# Display cleaned data
print(df_cleaned.head())

# Optional: save cleaned CSV
df_cleaned.to_csv('population_cleaned.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Population of India.csv")

# histogram for male population
plt.figure(figsize=(8,5))
plt.hist(df["Male"], bins=5)
plt.xlabel("Male Population")
plt.ylabel("Frequency")
plt.title("Histogram of Male Population Distribution")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Population of India.csv")

# Histogram for female population
plt.figure(figsize=(8,5))
plt.hist(df["Female"], bins=5, color="pink")
plt.xlabel("Female Population")
plt.ylabel("Frequency")
plt.title("Histogram of Female Population Distribution")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Population of India.csv")

# histogram for Male & Female combined
plt.figure(figsize=(12,6))
plt.bar(df["State/UT"], df["Male"], label="Male", alpha=0.7, color="blue")
plt.bar(df["State/UT"], df["Female"], label="Female", alpha=0.7, color="pink")

plt.xlabel("States/UT")
plt.ylabel("Population")
plt.title("Distribution of Male and Female Population Across States")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Population of India.csv')

# Sort by area for a cleaner chart
df_sorted = df.sort_values('Area[52] (km2)', ascending=False)

# Create the bar chart
plt.figure(figsize=(15,7))
plt.bar(df_sorted['State/UT'], df_sorted['Area[52] (km2)'], color='orange')
plt.xticks(rotation=90)  
plt.xlabel('State/UT')
plt.ylabel('Area (km²)')
plt.title('Area of Indian States and Union Territories')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Population of India.csv')

# Sort by density for a better visual
df_sorted = df.sort_values('Density (per km2)', ascending=False)

# Create horizontal bar chart
plt.figure(figsize=(12,10))
plt.barh(df_sorted['State/UT'], df_sorted['Density (per km2)'], color='green')
plt.xlabel('Density (per km²)')
plt.ylabel('State/UT')
plt.title('Population Density of Indian States and Union Territories')
plt.gca().invert_yaxis()  # Highest density on top
plt.tight_layout()
plt.show()
```
